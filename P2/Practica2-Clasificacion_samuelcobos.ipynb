```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report

# Manejo de datos faltantes
peruvian_blood_pressures = pd.read_csv('/Users/universidad/Desktop/Universidad/CURSO-2023-2024/BIO/PRACTICAS/P2/datasets/peruvian_blood_pressures.csv')
south_africa_chd = pd.read_csv('/Users/universidad/Desktop/Universidad/CURSO-2023-2024/BIO/PRACTICAS/P2/datasets/sourth_africa_chd.csv')

```

```python



systol_column = peruvian_blood_pressures['Systol']

systol_mean = np.mean(systol_column)
print("Media de la columna 'Systol':", systol_mean)

peruvian_blood_pressures.head()
```

```python

south_africa_chd.head()
```

```python
# Refactor data
ohe_famhist = OneHotEncoder(drop='if_binary').fit_transform(south_africa_chd[['famhist']])
south_africa_chd['famhist'] = ohe_famhist.toarray()

ohe_chd = OneHotEncoder(drop='if_binary').fit_transform(south_africa_chd[['chd']])
south_africa_chd['chd'] = ohe_chd.toarray()

peruvian_blood_pressures.dropna(inplace=True)
south_africa_chd.dropna(inplace=True)


X_peru = peruvian_blood_pressures.drop(['Systol', 'Diastol'], axis=1)
y_systol = peruvian_blood_pressures['Systol']
y_diastol = peruvian_blood_pressures['Diastol']

X_sa = south_africa_chd.drop('chd', axis=1)
y_chd = south_africa_chd['chd']


X_peru_train, X_peru_test, y_systol_train, y_systol_test, y_diastol_train, y_diastol_test = train_test_split(X_peru, y_systol, y_diastol, test_size=0.2, random_state=42)
X_sa_train, X_sa_test, y_chd_train, y_chd_test = train_test_split(X_sa, y_chd, test_size=0.2, random_state=42)

# # Escalado de características si es necesario
# scaler = StandardScaler()
# X_peru_train = scaler.fit_transform(X_peru_train)
# X_peru_test = scaler.transform(X_peru_test)

# X_sa_train = scaler.fit_transform(X_sa_train)
# X_sa_test = scaler.transform(X_sa_test)
```

```python
# Regresión
regressors = {
    'Linear Regression': LinearRegression(),
    'Decision Tree Regressor': DecisionTreeRegressor(),
    'Random Forest Regressor': RandomForestRegressor()
}

for name, model in regressors.items():
    model.fit(X_peru_train, y_systol_train)
    y_pred = model.predict(X_peru_test)
    mse = mean_squared_error(y_systol_test, y_pred)
    print(f'{name}: MSE = {mse}')

# Clasificación
classifiers = {
    'Logistic Regression': LogisticRegression(max_iter=100),
    'Decision Tree Classifier': DecisionTreeClassifier(),
    'Random Forest Classifier': RandomForestClassifier()
}

for name, model in classifiers.items():
    model.fit(X_sa_train, y_chd_train)
    y_pred = model.predict(X_sa_test)
    accuracy = accuracy_score(y_chd_test, y_pred)
    report = classification_report(y_chd_test, y_pred)
    print(f'{name}: Accuracy = {accuracy}')
    print(report)
```
